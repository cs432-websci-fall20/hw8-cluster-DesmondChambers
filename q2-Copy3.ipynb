```python
import fnmatch
import os
import re
from nltk.corpus import stopwords

all_tweets_dict = {}

for file in os.listdir('./tweets/'):
    try:
        #if fnmatch.fnmatch(file, '*.txt'):
        #print(file)
        my_file = open(f"./tweets/{file}", "r")
        content = my_file.read()
        globals()['{}'.format(file)] = content.split()
        my_file.close()
        #print(content_list)
        #for word in globals()[file]:
        all_tweets_dict['{}'.format(file)] = globals()['{}'.format(file)]
    except:
        continue
        
print(len(all_tweets_dict))
```

```python
clean_tweets_dict = {}
temp_list = []  
en_stops = set(stopwords.words('english'))

def clean_up(title, list): 
    for word in list:
        try:
            result = re.sub(r"http\S+", "", "{}".format(word.strip()))
            
            result = re.sub(r"@\S+", "", "{}".format(result.strip()))
            
            test = ''.join(letter for letter in result.lower() if 'a' <= letter <= 'z')
            
            if not len(test) <3 and not len(test) > 15 and test not in en_stops:
                temp_list.append(test)
               
            while("" in temp_list) : 
                temp_list.remove("") 
                
        except:
            continue
            
    #globals()['{}'.format(word)]
    #print(temp_list)
    final_list = [idx for idx in temp_list if not re.findall("[^\u0000-\u05C0\u2100-\u214F]+", idx)] 

    clean_tweets_dict['{}' .format(title)] = final_list
    #return temp_list


if __name__ == '__main__':
    for key, val in all_tweets_dict.items():   
        #clean_tweets_dict['{}'.format(key)] = clean_up(key, val)
        clean_up(key, val)
```

```python
def getwordcounts(title, tweets):
    wc = {}
    for word in tweets:
        #print(word)
        wc.setdefault(word, 0)
        wc[word] += 1
    #print(wc)
    return ("{}".format(title), wc)

apcount = {}
wordcounts = {}
for key, val in clean_tweets_dict.items():
    try:
        (title, wc) = getwordcounts(key, val)
        #print(val)
        wordcounts[title] = wc
        for (word, count) in wc.items():
            #print(word)
            apcount.setdefault(word, 0)
            if count > 1:
                apcount[word] += 1
            #print(apcount[word])
    except:
        print ('Failed to parse feed %s' % key)

wordlist = []
for (w, bc) in apcount.items():
    #print(bc)
    frac = float(bc) / len(clean_tweets_dict)
    #print(frac)
    if frac > 0.25 and frac < 0.39:
        wordlist.append(w)

#out = file('blogdata1.txt', 'w')
out = open('tweetdata.txt', 'w') 
out.write('Blog')
for word in wordlist:
    out.write('\t%s' % word)
    #print(word)
out.write('\n')
#phrase = dict(list(wordcounts.items())[:10])
for (blog, wc) in wordcounts.items():
#for (blog, wc) in phrase.items():
    #print (blog)
    out.write(blog)
    for word in wordlist:
        if word in wc:
            out.write('\t%d' % wc[word])
        else:
            out.write('\t0')
    out.write('\n')
print(len(wordlist))
```
